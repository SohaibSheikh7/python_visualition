```python

churn = pd.read_csv("Data.csv")
churn.head()
```

```python
#Number Of Customers Country wise
```

```python
plt.figure(figsize=(6,3))
plt.title("Number of Customers", fontsize=12)
plt.bar(x=churn['Country'].value_counts().index,
        height=churn.Country.value_counts().values)
```

```python
#Balance Dsitribution Customer wise
```

```python
plt.hist(x=churn['Balance'], bins=12,
         range=(25000, 225000))
plt.title("Distribution on Balance", fontsize=12)
```

```python
# Country Comparison France vs Spain Credit Score
```

```python
sample= churn.sample(n=200, random_state=20)
fig, ax = plt.subplots()

plt.title("France vs Spain", fontsize=14)
ax.scatter(x=sample[sample.Country == 'France']['CreditScore'], y=sample[sample.Country == 'France']['Age'])
ax.scatter(x=sample[sample.Country == 'Spain']['CreditScore'], y=sample[sample.Country == 'Spain']['Age'])
ax.legend(labels=['France','Spain'], loc='lower left', fontsize=12)
```

```python
# Country Comparison France vs Spain Estimated Score
```

```python
sample= churn.sample(n=200, random_state=20)
fig, ax = plt.subplots()

plt.title("France vs Spain", fontsize=14)
ax.scatter(x=sample[sample.Country == 'France']['EstimatedSalary'], y=sample[sample.Country == 'France']['Age'])
ax.scatter(x=sample[sample.Country == 'Spain']['EstimatedSalary'], y=sample[sample.Country == 'Spain']['Age'])
ax.legend(labels=['France','Spain'], loc='lower left', fontsize=12)
```

```python
# Subplot of Countries and Products
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True,
figsize=(6,4))
countries = churn.Country.value_counts()
products = churn.NumOfProducts.value_counts()
ax1.bar(x=countries.index, height=countries.values)
ax1.set_title("Countries", fontsize=12)
ax2.bar(x=products.index, height=products.values)
ax2.set_title("Number of Products", fontsize=12)
```
